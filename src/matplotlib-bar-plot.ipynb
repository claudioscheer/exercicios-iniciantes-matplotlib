### Example 1

#### Define the data that would be used.

```python
data = np.random.rand(5)
x = ["Claudio", "Luan", "Gabriel", "Djalma", "Greice"]
```

#### Create plot.

```python
plt.bar(x, data)
plt.show()
```

### Example 2

#### Define the data that would be used.

```python
data1 = [23, 85, 72, 43, 52]
data2 = [42, 35, 21, 16, 9]
width = 0.1
x = np.arange(5) + 1
```

#### Create plot.

```python
plt.bar(x, data1, label="Data 1", width=width)
plt.bar(x + width, data2, label="Data 2", width=width)
plt.bar(x + width * 2, data2, label="Data 3", width=width)
plt.ylabel("y")
plt.xlabel("x")
plt.title("Simple Example")
plt.legend()
plt.show()
```

### Example 3

#### Define the data that would be used.

```python
data1 = [23, 85, 72, 43, 52]
data2 = [42, 35, 21, 16, 9]
x = np.arange(5) + 1
```

#### Create plot.

```python
plt.bar(x, data1)
plt.bar(x, data2, bottom=data1)
plt.show()
```

### Example 4

#### Define the data that would be used.

```python
data = np.random.rand(5)
stds = np.random.rand(5)
x = np.arange(5)
```

#### Create plot.

```python
plt.errorbar(x, data, yerr=stds, fmt="-o", lw=1.3, ms=6, uplims=True, lolims=True)
plt.show()
```
