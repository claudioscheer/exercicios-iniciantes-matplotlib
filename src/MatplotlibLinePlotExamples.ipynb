### Example 1

#### Define the data that would be used.

```python
x = np.linspace(-1, 1, 10)
y1 = 2 * x + 1
y2 = 2**x + 1
```

#### Create plot.

```python
figure = plt.figure()
plt.plot(x, y1, label="Data 1")
plt.plot(x, y2, label="Data 2", linestyle="--")
plt.ylabel("y")
plt.xlabel("x")
plt.title("Simple Example 1")
plt.legend()
plt.show()
```

### Example 2

#### Define the data that would be used.

```python
x = np.linspace(0, 10000, 10)
y1 = np.random.rand(10) * 100
y2 = np.random.rand(10) * 100
y3 = np.random.rand(10) * 100
```

#### Create plot.

```python
figure = plt.figure()
plt.plot(x, y1, label="gcc")
plt.plot(x, y2, label="g++")
plt.plot(x, y3, label="clang", marker="o")
plt.ylabel("Build Time")
plt.xlabel("Executions")
plt.title("Simple Example 2")
plt.legend()
plt.show()
```
